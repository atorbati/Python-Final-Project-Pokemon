### Grading scores

All cards graded by PSA will be given a score depending on the quality of the printed card. Some cards can be miscuts/off-center which are manufacturing errors that slipped through quality control(12). While some people grade these cards, they are not included in this report as they are labeled as (MC) next to the grade (eg. Jirachi Holographic 9(MC)) and they are also so rare that there is not enough data per card to be analyzed(13).

Terms such as Gem Mint, Mint, Near Mint, etc. are ignored in this report as the numerical grade is used more widely by data aggregators and sellers. Outside of miscuts, grades usually follow the rubric provided by PSA:

**PSA 10**

A PSA Gem Mint 10 card is a virtually perfect card.
Attributes include four perfectly sharp corners, sharp focus and full original gloss. A PSA Gem Mint 10 card must be free of staining of any kind, but an allowance may be made for a slight printing imperfection, if it doesn't impair the overall appeal of the card. The image must be centered on the card within a tolerance not to exceed approximately 55/45 percent on the front, and 75/25 percent on the reverse.

**PSA 9**

A PSA Mint 9 is a superb condition card that exhibits only one of the following minor flaws: a very slight wax stain on reverse, a minor printing imperfection or slightly off white borders. Centering must be approximately 60/40 or better on the front and 90/10 or better on the reverse.

**PSA 8**

A PSA NM-MT 8 is a super high-end card that appears Mint 9 at first glance, but upon closer inspection, the card can exhibit the following: a very slight wax stain on reverse, slightest fraying at one or two corners, a minor printing imperfection, and/or slightly off-white borders. Centering must be approximately 65/35 or better on the front and 90/10 or better on the reverse.

**PSA 7**

A PSA NM 7 is a card with just a slight surface wear visible upon close inspection. There may be slight fraying on some corners. Picture focus may be slightly out-of register. A minor printing blemish is acceptable. Slight wax staining is acceptable on the back of the card only. Most of the original gloss is retained. Centering must be approximately 70/30 or better on the front and 90/10 or better on the back.

**PSA 6**

A PSA 6 card may have visible surface wear or a printing defect which does not detract from its overall appeal. A very light scratch may be detected only upon close inspection. Corners may have slightly graduated fraying. Picture focus may be slightly out-of-register. Card may show some loss of original gloss, may have minor wax stain on reverse, may exhibit very slight notching on edges and may also show some off-whiteness on borders. Centering must be 80/20 or better on the front and 90/10 or better on the reverse.

**PSA 5**

On PSA 5 cards, very minor rounding of the corners is becoming evident. Surface wear or printing defects are more visible. There may be minor chipping on edges. Loss of original gloss will be more apparent. Focus of picture may be slightly out-of-register. Several light scratches may be visible upon close inspection, but do not detract from the appeal of the card. Card may show some off-whiteness of borders. Centering must be 85/15 or better on the front and 90/10 or better on the back.

**PSA 4**

A PSA 4 card's corners may be slightly rounded. Surface wear is noticeable but modest. The card may have light scuffing or light scratches. Some original gloss will be retained. Borders may be slightly off-white. A light crease may be visible. Centering must be 85/15 or better on the front and 90/10 or better on the back.

**PSA 3**

A PSA 3 card reveals some rounding of the corners, though not extreme. Some surface wear will be apparent, along with possible light scuffing or light scratches. Focus may be somewhat off-register and edges may exhibit noticeable wear. Much, but not all, of the card's original gloss will be lost. Borders may be somewhat yellowed and/or discolored. A crease may be visible. Printing defects are possible. Slight stain may show on obverse and wax staining on reverse may be more prominent. Centering must be 90/10 or better on the front and back.

**PSA 2**

A PSA 2 card's corners show accelerated rounding and surface wear is starting to become obvious. A good card may have scratching, scuffing, light staining, or chipping of enamel on obverse. There may be several creases. Original gloss may be completely absent. Card may show considerable discoloration. Centering must be 90/10 or better on the front and back.

**PSA 1**

A PSA 1 will exhibit many of the same qualities of a PSA 1.5 but the defects may have advanced to such a serious stage that the eye appeal of the card has nearly vanished in its entirety. A Poor card may be missing one or two small pieces, exhibit major creasing that nearly breaks through all the layers of cardboard or it may contain extreme discoloration or dirtiness throughout that may make it difficult to identify the issue or content of the card on either the front or back. A card of this nature may also show noticeable warping or another type of destructive defect.


NOTE: There are other grades such as Authentic, Altered Authentic, and Half-Points but they are very uncommon and not included in most datasets reporting graded card prices.


### Grading Cost and Upcharging

The cost of grading is dependent on two variables; the value of the ungraded card and the estimated value of the card after grading(14).

PSA first charges the customer based on the value of the ungraded card:

**“Value” - Max card value 500 USD - cost: 27.99 per card**

**“Value Max” - Max card value 1000 USD - cost: 59.99 per card**

**“Regular” - Max card value 1500 USD - cost: 79.99 per card**

**“Express” - Max card value 2500 USD - cost: 149.99 per card**

**“Super Express” - Max card value 5000 USD - cost: 299.99 per card**

**“Walk-Through” - Max card value 10000 USD - cost: 599.99 per card**


Upcharging occurs when the estimated value of the graded card; if a graded card goes over the max card value the customer must pay the difference between the different grading options.

PSA does **not** provide information about this anywhere on their website - the customer does not know about this upcharge until after the grading occurs and before the graded card is shipped to them. This usually is displayed as a fee right next to the shipping cost (Figure 2). The only sources of this phenomenon are on forums, social media, and hobby websites(15).

```python
# lets get started
import io, os
from pathlib import Path
import warnings

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import MatplotlibDeprecationWarning

# remove the annoying warnings
warnings.filterwarnings("ignore", category=MatplotlibDeprecationWarning)

RAW_URL = (
    "https://raw.githubusercontent.com/atorbati/Python-Final-Project-Pokemon/"
    "863d0f27b685a24e35dceedaf77e56bb986736d7/FINAL_PYTHON_II_PROJECTv2.csv"
)

OUT_DIR = Path("./outputs"); OUT_DIR.mkdir(exist_ok=True)

# Cost of shipping and individual booster pack of cards via psa and pitt respectively 
PACK_PRICE_USD   = 8.30
SHIP_ONE_WAY_LOW = 19.99     # ≤ $2000 raw value
SHIP_ONE_WAY_HIGH= 34.99     #  > $2000 raw value
VALUE_THRESHOLD  = 2000.0
TWO_WAY          = 2

# Numeric columns
NUMERIC_COLS = [
    "ungraded_price","PSA_nine","PSA_ten",
    "PSA_upcharge_10","PSA_upcharge_9","PSA_grade_price",
    "psa_difference","one_in_x_cards_rarity","one_in_x_packs_individual",
    "Card_population_PSA_9","Card_population_PSA_10","Card_population_total",
    "Profit_per_card_PSA_9","Profit_per_card_PSA_10",
]

# Correlation
CORR_COLS = [
    "ungraded_price","PSA_nine","PSA_ten",
    "Card_population_PSA_9","Card_population_PSA_10","Card_population_total",
    "Profit_per_card_PSA_9","Profit_per_card_PSA_10",
]

# save those figures just in case
SAVE_FIGS = True
FIG_DPI   = 150

def _savefig(name: str):
    if SAVE_FIGS:
        plt.savefig(OUT_DIR / f"{name}.png", dpi=FIG_DPI, bbox_inches="tight")


def load_df(url: str) -> pd.DataFrame:
    try:
        return pd.read_csv(url, index_col=0)
    except Exception:
        import requests, certifi
        os.environ["SSL_CERT_FILE"] = certifi.where()
        r = requests.get(url, timeout=30, verify=certifi.where())
        r.raise_for_status()
        return pd.read_csv(io.StringIO(r.text), index_col=0)


# cleaning better than my house
def clean_types(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df.columns = [c.strip() for c in df.columns]

    price_like_cols = [
        "ungraded_price","PSA_nine","PSA_ten",
        "PSA_grade_price","PSA_upcharge_9","PSA_upcharge_10",
    ]

    def to_number(series):
        s = series.astype(str)
        s = (s.str.replace("−", "-", regex=False)
               .str.replace("–", "-", regex=False)
               .str.replace(r"\(([^)]+)\)", r"-\1", regex=True)
               .str.replace(r"[\$,]", "", regex=True)
               .str.strip())
        return pd.to_numeric(s, errors="coerce")

    for col in price_like_cols:
        if col in df.columns:
            df[col] = to_number(df[col])

    for col in NUMERIC_COLS:
        if col in df.columns and col not in price_like_cols:
            df[col] = pd.to_numeric(df[col], errors="coerce")

    return df


def add_grading_shipping(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    if "ungraded_price" in df.columns:
        df["two_way_ship_for_grading"] = np.where(
            df["ungraded_price"] > VALUE_THRESHOLD,
            TWO_WAY * SHIP_ONE_WAY_HIGH,
            TWO_WAY * SHIP_ONE_WAY_LOW,
        )
    else:
        df["two_way_ship_for_grading"] = np.nan
    return df

def add_total_costs_and_margins(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    if {"ungraded_price","PSA_grade_price","PSA_upcharge_9","two_way_ship_for_grading"}.issubset(df.columns):
        df["total_cost_psa9"] = (
            df["ungraded_price"].fillna(0)
          + df["PSA_grade_price"].fillna(0)
          + df["PSA_upcharge_9"].fillna(0)
          + df["two_way_ship_for_grading"].fillna(0)
        )

    if {"ungraded_price","PSA_grade_price","PSA_upcharge_10","two_way_ship_for_grading"}.issubset(df.columns):
        df["total_cost_psa10"] = (
            df["ungraded_price"].fillna(0)
          + df["PSA_grade_price"].fillna(0)
          + df["PSA_upcharge_10"].fillna(0)
          + df["two_way_ship_for_grading"].fillna(0)
        )

    if {"PSA_nine","total_cost_psa9"}.issubset(df.columns):
        denom = df["total_cost_psa9"].replace(0, np.nan)
        df["margin_pct_psa9_self_grade"] = (df["PSA_nine"] - df["total_cost_psa9"]) / denom

    if {"PSA_ten","total_cost_psa10"}.issubset(df.columns):
        denom = df["total_cost_psa10"].replace(0, np.nan)
        df["margin_pct_psa10_self_grade"] = (df["PSA_ten"] - df["total_cost_psa10"]) / denom

    return df

def add_expected_pull_cost(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    if "one_in_x_packs_individual" in df.columns:
        df["expected_pull_cost"] = df["one_in_x_packs_individual"] * PACK_PRICE_USD
        df["cheaper_method_raw"] = np.where(
            df["expected_pull_cost"] < df.get("ungraded_price", np.nan),
            "Pull from packs", "Buy ungraded online"
        )
    return df

def rarity_distribution(df: pd.DataFrame) -> pd.DataFrame:
    if "card_rarity" not in df.columns: return pd.DataFrame()
    # exclude NaN so it doesn’t appear as a category
    return (
        df.loc[df["card_rarity"].notna(), "card_rarity"]
          .value_counts()
          .rename_axis("card_rarity")
          .reset_index(name="count")
    )

def correlation_matrix(df: pd.DataFrame) -> pd.DataFrame:
    cols = [c for c in CORR_COLS if c in df.columns]
    if not cols: return pd.DataFrame()
    return df[cols].corr(method="pearson")

def _find_name_column(df: pd.DataFrame):
    candidates = ["card_name","Card_Name","Card","name","Name","title"]
    for c in candidates:
        if c in df.columns:
            return c
    return None


# needed to make it backwards compatable so it wouldnt give me other warnings or errors
def boxplot_with_labels(data, labels, **kwargs):
    try:
        return plt.boxplot(data, tick_labels=labels, **kwargs)  
    except TypeError:
        return plt.boxplot(data, labels=labels, **kwargs)       

def _fd_bins(x: np.ndarray, max_bins: int = 80) -> int:
    x = x[~np.isnan(x)]
    if x.size < 2: return 10
    iqr = np.subtract(*np.percentile(x, [75, 25]))
    if iqr == 0: return min(max_bins, max(10, int(np.sqrt(x.size))))
    bw = 2 * iqr / (x.size ** (1/3))
    if bw <= 0: return 30
    b = int(np.ceil((x.max() - x.min()) / bw))
    return max(10, min(max_bins, b))


# lots o charts
def plot_rarity_bar(rarity_counts: pd.DataFrame) -> None:
    if rarity_counts.empty: return
    plt.figure()
    plt.bar(rarity_counts["card_rarity"].astype(str), rarity_counts["count"])
    plt.title("Rarity Distribution (Counts)")
    plt.xlabel("Card Rarity"); plt.ylabel("Count")
    plt.xticks(rotation=45, ha="right")
    plt.tight_layout(); _savefig("rarity_bar"); plt.show()

def plot_hist_pull_rates(df: pd.DataFrame) -> None:
    col = "one_in_x_packs_individual"
    if col not in df.columns: return
    vals = pd.to_numeric(df[col], errors="coerce").dropna().values
    if len(vals) == 0: return
    plt.figure()
    plt.hist(vals, bins=_fd_bins(vals))
    plt.title("Histogram: Packs Needed to Pull a Card (1 in X)")
    plt.xlabel("Packs Needed (1 in X)"); plt.ylabel("Frequency")
    plt.tight_layout(); _savefig("hist_pull_rates"); plt.show()

def plot_box_prices(df: pd.DataFrame, quantile=0.99, hard_max=4000) -> None:
    """Readable box plot. Uses robust upper bound (quantile), capped at hard_max (default 4000)."""
    cols = [c for c in ["ungraded_price","PSA_nine","PSA_ten"] if c in df.columns]
    if not cols: print("No price columns found."); return
    arrays = [pd.to_numeric(df[c], errors="coerce").dropna().values for c in cols]
    if sum(a.size for a in arrays) == 0: print("No finite price data."); return

    q = np.nanpercentile(np.concatenate(arrays), quantile*100)
    y_max = float(min(q, hard_max)) if hard_max is not None else float(q)
    y_min = 0.0

    above = {c: int((pd.to_numeric(df[c], errors="coerce") > y_max).sum()) for c in cols}

    plt.figure()
    boxplot_with_labels(arrays, cols, showfliers=False)
    plt.title("Box Plots: Price Distributions (Ungraded vs PSA 9 vs PSA 10)")
    plt.ylabel("USD")
    plt.ylim(y_min, y_max)
    plt.tight_layout(); _savefig("box_prices"); plt.show()
    print(f"(Info) y-axis 0–{y_max:,.0f} (from {int(quantile*100)}th pct; capped at {hard_max}). Above-cap counts:", above)

def plot_box_profit_margins(df: pd.DataFrame) -> None:
    cols = [c for c in ["margin_pct_psa9_self_grade","margin_pct_psa10_self_grade"] if c in df.columns]
    if not cols: return
    arrays = [pd.to_numeric(df[c], errors="coerce").dropna().values for c in cols]
    plt.figure()
    boxplot_with_labels(arrays, cols, showfliers=False)
    plt.title("Box Plots: Profit Margin (%) Distributions")
    plt.ylabel("Margin (fraction)")
    plt.tight_layout(); _savefig("box_profit_margins"); plt.show()

def plot_hist_prices(df: pd.DataFrame, log_x=False, x_upper_quantile=0.995):
    for col in ["ungraded_price","PSA_nine","PSA_ten"]:
        if col in df.columns and df[col].notna().any():
            x = pd.to_numeric(df[col], errors="coerce").dropna().values
            if x.size == 0: 
                continue
            xmax = np.quantile(x, x_upper_quantile)
            x_plot = x[x <= xmax]

            plt.figure()
            plt.hist(x_plot, bins=_fd_bins(x_plot))
            plt.title(f"Histogram: {col}")
            plt.xlabel("USD"); plt.ylabel("Count")
            if log_x: plt.xscale("log")
            plt.tight_layout(); _savefig(f"hist_{col}"); plt.show()

            if (x > xmax).any():
                print(f"(Info) {col}: {np.sum(x > xmax)} values > {xmax:,.2f} trimmed for readability.")

def plot_scatter_ungraded_vs_psa10(df: pd.DataFrame) -> None:
    if not {"ungraded_price","PSA_ten"}.issubset(df.columns): return
    x = pd.to_numeric(df["ungraded_price"], errors="coerce").values
    y = pd.to_numeric(df["PSA_ten"], errors="coerce").values
    mask = np.isfinite(x) & np.isfinite(y)
    if mask.sum() == 0: return
    plt.figure()
    plt.scatter(x[mask], y[mask])
    plt.title("Scatter: Ungraded Price vs PSA 10 Price")
    plt.xlabel("Ungraded Price (USD)"); plt.ylabel("PSA 10 Price (USD)")
    plt.tight_layout(); _savefig("scatter_ungraded_vs_psa10"); plt.show()

def plot_heatmap_corr(corr: pd.DataFrame) -> None:
    if corr.empty: return
    plt.figure()
    plt.imshow(corr, aspect="auto")
    plt.colorbar()
    plt.title("Correlation Matrix (Pearson r)")
    plt.xticks(ticks=range(len(corr.columns)), labels=corr.columns, rotation=45, ha="right")
    plt.yticks(ticks=range(len(corr.index)), labels=corr.index)
    plt.tight_layout(); _savefig("heatmap_correlation"); plt.show()

def plot_bar_top_profit(df: pd.DataFrame, n: int = 15) -> None:
    metric = None
    if "margin_pct_psa10_self_grade" in df.columns and df["margin_pct_psa10_self_grade"].notna().any():
        metric = "margin_pct_psa10_self_grade"
    elif "margin_pct_psa9_self_grade" in df.columns and df["margin_pct_psa9_self_grade"].notna().any():
        metric = "margin_pct_psa9_self_grade"
    if metric is None:
        print("No profit metric available."); return

    name_col = _find_name_column(df)
    cols = [metric] + ([name_col] if name_col else [])
    subset = df[cols].dropna(subset=[metric]).sort_values(metric, ascending=False).head(n).copy()
    if subset.empty: 
        print("No rows to plot."); return

    labels = subset[name_col].astype(str).values if name_col else subset.index.astype(str).values
    plt.figure(figsize=(8, 6))
    plt.barh(labels, subset[metric].values)
    plt.gca().invert_yaxis()
    plt.title(f"Top {len(subset)} Cards by {metric}")
    plt.xlabel("Margin (fraction)"); plt.ylabel("Card")
    plt.tight_layout(); _savefig("bar_top_profit"); plt.show()
        


# save this so i dont lose it all right before i submit...
def save_summary(df: pd.DataFrame) -> Path:
    cols = [
        "card_rarity","ungraded_price","PSA_nine","PSA_ten",
        "total_cost_psa9","margin_pct_psa9_self_grade",
        "total_cost_psa10","margin_pct_psa10_self_grade",
        "one_in_x_packs_individual","expected_pull_cost","cheaper_method_raw",
    ]
    # idk about name column my brain is fried
    name_col = _find_name_column(df)
    if name_col: cols = [name_col] + cols

    present = [c for c in cols if c in df.columns]
    out = OUT_DIR / "prismatic_analysis_summary.csv"
    df[present].to_csv(out, index=False)
    return out


# almost done...
def main():
    df = load_df(RAW_URL)
    df = clean_types(df)
    df = add_grading_shipping(df)
    df = add_total_costs_and_margins(df)
    df = add_expected_pull_cost(df)

    rarity = rarity_distribution(df)
    corr   = correlation_matrix(df)

    plot_rarity_bar(rarity)
    plot_hist_pull_rates(df)
    plot_box_prices(df, quantile=0.99, hard_max=4000)  # enforces readable 0–~cap
    plot_box_profit_margins(df)
    plot_hist_prices(df, log_x=False, x_upper_quantile=0.995)
    plot_scatter_ungraded_vs_psa10(df)
    plot_heatmap_corr(corr)
    plot_bar_top_profit(df, n=15)

    out_csv = save_summary(df)

# Run
main()
```

## Results 

Looking at the charts, it is clear that investing in Pokémon cards **can** be profitable, but only if you are selective about which cards you buy and how you approach the market. The second-hand Pokémon card market (both graded and ungraded) should be evaluated on a short-term basis; long-term value forecasts remain highly speculative. Since the market isn’t regulated, it’s easy for prices to be manipulated or inflated by sellers and hype. Because of this, please take all findings with a grain of salt, as the market can shift quickly.

The box plots and histograms show that most cards hold modest value; ungraded cards usually sell for under $50, and even graded cards, such as PSA 9s and PSA 10s, tend to remain under $100. A few rare or high-grade cards reach prices in the hundreds or even thousands, yet those are outliers rather than the norm. The correlation heatmap confirms that ungraded, PSA 9, and PSA 10 prices move closely together, showing that value depends mainly on card condition and collector demand rather than random market hype. Scarcity plays a role in price but does not guarantee high value; a card can be rare but still unwanted if collectors are not seeking it.

The profit margin results suggest that grading can sometimes pay off, especially for PSA 10 cards, but overall returns remain limited once grading and shipping costs are included. Many cards show little to no profit after fees, and some even produce losses. The pull-rate histograms further explain this outcome; truly rare cards appear infrequently, sometimes only once in dozens or hundreds of booster packs. This makes opening sealed packs more of a gamble than a consistent investment. For most cards, the expected pull cost exceeds the potential resale price, meaning that chasing profit through pack openings is not financially wise.

Buying cards directly from the secondary market appears to be the more practical and responsible strategy. Purchasing ungraded cards that seem undervalued, or selecting PSA-graded cards that show strong collector demand, allows for more control over spending and reduces the influence of chance. For hobbyists who enjoy the excitement of opening packs, the activity still has value, but it should be viewed as entertainment rather than a business opportunity.

## Conclusion

In conclusion, Pokémon cards can generate profit under specific circumstances, but success depends on informed decision-making and realistic expectations. The evidence shows that buying individual cards, especially those with stable demand and solid grading potential, is a safer and more sustainable approach than relying on booster pack openings. This strategy balances enjoyment of the hobby with financial responsibility and long-term value.
